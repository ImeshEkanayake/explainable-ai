```python
import numpy as np
import pandas as pd
from textblob import TextBlob
```

```python
Data = pd.read_csv('Projects.csv')
```

```python
Data.columns
```

```python
def polarity_and_subjectivity_of_text(Data,column_name):
    Data[column_name+"_polarity"] = 0
    Data[column_name+"_subjectivity"] = 0
    for i in range(len(Data)):#len(Data)
        
        if(pd.isnull(Data[column_name][i])):
            Data[column_name+"_polarity"][i] = 0
            Data[column_name+"_subjectivity"][i] = 0
        else:
            analysis = TextBlob(Data[column_name][i])
            pol=analysis.sentiment.polarity
            sen=analysis.sentiment.subjectivity
            Data[column_name+"_polarity"][i] = str(pol)
            Data[column_name+"_subjectivity"][i] = str(sen)


        if (i%5000)==0:  print(i,"Done")
            
    return Data
```

```python
polarity_and_subjectivity_of_text(Data,'Project Short Description')
```

```python
polarity_and_subjectivity_of_text(Data,'Project Essay')
```

```python
polarity_and_subjectivity_of_text(Data,'Project Need Statement')
```
