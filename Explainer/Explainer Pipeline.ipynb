```python
Data = pd.read_csv('X_test_tosend.csv') 
```

```python
def NearestN_N_Neighbours(data,pointA,NumberOfinstances):
    mean_of_data=data.mean()
    std_of_data=data.std()
    normlized_data=(data-mean_of_data)/std_of_data
    normlized_pointA=(pointA-mean_of_data)/std_of_data
    local_distance= (normlized_data - normlized_pointA)**2
    data["distance"]=local_distance.sum(axis=1)
    data=data.sort_values("distance")
    data_nearest_K=data.head(NumberOfinstances)
    return data_nearest_K
```

```python
pointA=Data.loc[0,:]
NearestN_N_Neighbours(Data,pointA,100)
```
