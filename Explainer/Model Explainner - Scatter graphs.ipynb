```python
import os
import random 
import datetime
import numpy as np
import pandas as pd
from sklearn import tree
from datetime import timedelta
import matplotlib.pyplot as plt
from sklearn.impute import KNNImputer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import plot_precision_recall_curve
import pickle
```

```python
objects = []
with (open("Dict_CatBoostClassifier_with_Explanation_22_2_2021_01_F", "rb")) as openfile:
    while True:
        try:
            objects.append(pickle.load(openfile))
        except EOFError:
            break
```

```python
Output = objects[0]
```

```python
Output.keys()
```

```python
K=100
dict_top = dict()
top_ats=15
count=0
cols=[]
dates=[]

Lime_vals=[]
Data_vals=[]
for date_index in range (3,len(Output['Time_points']),4):
    print("===============================",Output['Time_points'][date_index],"=================================")
    As_of_top_K = Output["Explainer_df_Top_K"][date_index].apply(pd.to_numeric, errors='ignore')
    As_of_bottom_K = Output["Explainer_df_Bottom_K"][date_index].apply(pd.to_numeric, errors='ignore')
    As_of_mid_K = Output["Explainer_df_Mid_K"][date_index].apply(pd.to_numeric, errors='ignore')
    
    As_of_Full_K = [As_of_top_K, As_of_mid_K, As_of_bottom_K]
    As_of_Full_K = pd.concat(As_of_Full_K)
    
    As_of_top_K=As_of_Full_K
    
    As_of_top_K_np = As_of_top_K.to_numpy()
    As_of_top_K_np[As_of_top_K_np < 0] = 0
    As_of_top_K_pos_df = pd.DataFrame(data=As_of_top_K_np, columns=As_of_top_K.columns)  # 1st row as the column names
    
    As_of_top_K_pos_df=As_of_top_K_pos_df.T
    Normalized_As_of_top_K_pos_df=(As_of_top_K_pos_df-As_of_top_K_pos_df.mean())/(As_of_top_K_pos_df.max()-As_of_top_K_pos_df.min())
    Normalized_As_of_top_K_pos_df=Normalized_As_of_top_K_pos_df.T
    
    important_columns_df = Normalized_As_of_top_K_pos_df.dropna(axis=1, how='all')
    
    Original_data = Output["Testing_data"][count]
    dates+=[Output['Time_points'][date_index]]
    dict_top[Output['Time_points'][date_index]]=dict()
    
    col_name='Project Cost'
    
    print(Output["Testing_data"][count].shape,list(As_of_top_K.index)[-1])
    count+=1
    plt.scatter(Original_data.loc[list(As_of_top_K.index),[col_name]],important_columns_df[col_name])
    plt.xscale('log')
    plt.xlabel(col_name)
    plt.ylabel("LIME Normalized importance value")
    
    for col_name in important_columns_df.columns:
        if col_name not in cols: 
            cols+=[col_name]

        co = np.corrcoef(np.array(Original_data.loc[list(As_of_top_K.index),[col_name]])[:,0],
                         np.array(important_columns_df[col_name]))[0,1]
#         print(co)
        if pd.notna(co):
            dict_top[Output['Time_points'][date_index]][col_name]=co

    

df_Top = pd.DataFrame(columns=dates)
for col_name in cols:
    df_Top.loc[col_name] = 0.0
for d in dict_top:
    for c in dict_top[d]:
        df_Top.at[c, d] = dict_top[d][c]
df_Top_transposed = df_Top.T
describe_top_df = df_Top_transposed.describe().T
describe_top_df= describe_top_df.sort_values(by=['mean'], ascending=False)

fig, axs = plt.subplots(5, 2,figsize=(15,25))
        
for i in range (5):
    axs[i, 0].scatter(df_Top_transposed.index, df_Top_transposed[describe_top_df.index[i]])
    axs[i, 0].title.set_text("Top_K, Middle_K and Bottom_K "+describe_top_df.index[i])
#     axs[i, 0].ylabel("LIME Normalized importance value")
    plt.ylabel("LIME Normalized importance value")

for i in range (1,6):
    axs[i-1, 1].scatter(df_Top_transposed.index, df_Top_transposed[describe_top_df.index[-i]])
    axs[i-1, 1].title.set_text("Top_K, Middle_K and Bottom_K "+describe_top_df.index[-i]) 
#     axs[i-1, 1].ylabel("LIME Normalized importance value")
    plt.ylabel("LIME Normalized importance value")
            

plt.show()
```
